```python
import pandas as pd
import re
```

```python
df = pd.read_csv('calendar.csv')
df
```

Copy **subject** field in **calendar.csv** to **Title** metadata field in posts

```python
for index, row in df.iterrows():
    if type(row.Filename) == str:
        filename = 'posts/{}'.format(row.Filename)
        #print(filename)
            
        f = open(filename,'r')
        filedata = f.read()
        
        m = re.search('Title: .*\n', filedata)
        if m:
            old_title = m.group()
        else:
            old_title = None
        f.close()
        
        if old_title:
            new_title = 'Title: {}\n'.format(row.Subject)
            filedata = filedata.replace(old_title, new_title)
            #print('|', old_title, '|')
            #print('|', new_title, '|')
            #print()

            f = open(filename,'w')
            f.write(filedata)
            f.close()    
```

Copy **Start Date** and **Start Time** fields in **calendar.csv** to **Date** metadata field in posts

```python
for index, row in df.iterrows():
    if type(row.Filename) == str:
        filename = 'posts/{}'.format(row.Filename)
        #print(filename)
            
        f = open(filename,'r')
        filedata = f.read()
        
        m = re.search('Date: .*\n', filedata)
        if m:
            old_date = m.group()
        else:
            old_date = None
        f.close()
        
        if old_date:
            new_date = 'Date: {} {}\n'.format(row['Start Date'], row['Start Time'])
            filedata = filedata.replace(old_date, new_date)
            #print('|', old_date, '|')
            #print('|', new_date, '|')
            #print()

            f = open(filename,'w')
            f.write(filedata)
            f.close()    
```
